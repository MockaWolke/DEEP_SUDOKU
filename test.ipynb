```python
start_slices = []
size = [-1, 3, 3, -1]

for i in range(3):
    for y in range(3):
        start_slices.append([0,i,y,0])

start_slices = tf.convert_to_tensor(start_slices)
size = tf.convert_to_tensor(size)

MSE = tf.keras.losses.MeanSquaredError()
```

```python
@tf.function
def some(x):
    
    reg_loss = tf.convert_to_tensor(0, tf.float32)
    
    for start in start_slices:
    
        reg = tf.slice(x, start,size)
        reg = tf.reduce_sum(x, axis = [1, 2])
        reg_loss += MSE(reg, tf.ones_like(reg))
        
    return reg_loss
some(x)
```
